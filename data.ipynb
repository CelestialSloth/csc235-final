```python
import pandas as pd
```

```python
converter_df = pd.read_csv('./final_data/zip_to_uhf34.csv')
trees_raw_df = pd.read_csv('./final_data/tree_data/census_2015/street_tree_census_2015.csv')
ozone_df = pd.read_csv('./final_data/environment_data/ozone.csv')
pm_df = pd.read_csv('./final_data/environment_data/PM.csv')
temp_df = pd.read_csv('./final_data/environment_data/temperature.csv')

```

```python
ozone_df = pd.read_csv('./final_data/environment_data/ozone.csv')
pm_df = pd.read_csv('./final_data/environment_data/PM.csv')
temp_df = pd.read_csv('./final_data/environment_data/temperature.csv')

```

```python
ozone_borough_df = ozone_df.query('GeoRank==1 & TimePeriod > "2√"').drop(columns=['TimePeriod','GeoType','GeoRank','10th percentile ppb','90th percentile ppb'])
ozone_borough_df = ozone_borough_df.groupby('Geography').mean()

ozone_UHF_df = ozone_df.query('GeoRank==3 & TimePeriod > "2√"').drop(columns=['TimePeriod','GeoType','GeoRank','10th percentile ppb','90th percentile ppb'])
ozone_UHF_group = ozone_UHF_df.groupby('GeoID')
ozone_UHF_df = ozone_UHF_group[['Mean ppb']].mean()
ozone_UHF_df['Geography'] = ozone_UHF_group[['Geography']].first()

pm_borough_df = pm_df.query('GeoRank==1 & TimePeriod < "B"').drop(columns=['TimePeriod','GeoType','GeoRank','10th percentile mcg/m3','90th percentile mcg/m3'])
pm_borough_df = pm_borough_df.groupby('Geography').mean()

pm_UHF_df = pm_df.query('GeoRank==3 & TimePeriod < "B"').drop(columns=['TimePeriod','GeoType','GeoRank','10th percentile mcg/m3','90th percentile mcg/m3'])
pm_UHF_group = pm_UHF_df.groupby('GeoID')
pm_UHF_df = pm_UHF_group[['Mean mcg/m3']].mean()
pm_UHF_df['Geography'] = pm_UHF_group[['Geography']].first()
pm_UHF_df.reset_index()


temp_borough_df = temp_df.query('GeoRank==1').drop(columns=['TimePeriod','GeoType','GeoRank'])
temp_borough_df = temp_borough_df.groupby('Geography').mean()

temp_UHF42_df = temp_df.query('GeoRank==4').drop(columns=['TimePeriod','GeoType','GeoRank'])
temp_UHF42_df['GeoID'].replace(305,305307, inplace=True)
temp_UHF42_df['GeoID'].replace(307,305307, inplace=True)
temp_UHF42_df['GeoID'].replace(306,306308, inplace=True)
temp_UHF42_df['GeoID'].replace(308,306308, inplace=True)
temp_UHF42_df['GeoID'].replace(309,309310, inplace=True)
temp_UHF42_df['GeoID'].replace(404,404406, inplace=True)
temp_UHF42_df['GeoID'].replace(406,404406, inplace=True)
temp_UHF42_df['GeoID'].replace(501,501502, inplace=True)
temp_UHF42_df['GeoID'].replace(502,501502, inplace=True)
temp_UHF42_df['GeoID'].replace(503,503504, inplace=True)
temp_UHF42_df['GeoID'].replace(504,503504, inplace=True)
temp_UHF42_df['GeoID'].replace(105,105106107, inplace=True)
temp_UHF42_df['GeoID'].replace(106,105106107, inplace=True)
temp_UHF42_df['GeoID'].replace(107,105106107, inplace=True)
temp_UHF42_group = temp_UHF42_df.groupby('GeoID')
temp_UHF_df = temp_UHF42_group[['Degrees Fahrenheit']].mean()
temp_UHF_df['Geography'] = temp_UHF42_group[['Geography']].first()
temp_UHF_df.reset_index()
temp_UHF_group = temp_UHF_df.groupby('GeoID')
temp_UHF_df = temp_UHF_group['Degrees Fahrenheit'].mean()
temp_UHF_df['Geography'] = temp_UHF_group['Geography'].first()
```

```python
combined_env_UHF_df = pd.merge(temp_UHF_df, pm_UHF_df, left_on="GeoID", right_on="GeoID",how="inner", sort=False)
combined_env_UHF_df = pd.merge(combined_env_UHF_df, ozone_UHF_df, left_on="GeoID", right_on="GeoID",how="inner", sort=False )
combined_env_UHF_df['UHF34 Neighborhood Name'] = combined_env_UHF_df['Geography_x']
combined_env_UHF_df = combined_env_UHF_df.drop(columns=['Geography_x', 'Geography_y'])

combined_env_borough_df = pd.merge(temp_borough_df, pm_borough_df, left_on="Geography", right_on="Geography",how="inner", sort=False)
combined_env_borough_df = pd.merge(combined_env_borough_df, ozone_borough_df, left_on="Geography", right_on="Geography",how="inner", sort=False )
combined_env_borough_df=combined_env_borough_df.reset_index()
combined_env_borough_df['Borough'] = combined_env_borough_df['Geography']
combined_env_borough_df = combined_env_borough_df.drop(columns=['Geography', 'GeoID_x', 'GeoID_y', 'GeoID'])
combined_env_borough_df
```

```python
trees_raw_df['zipcode'].replace(83,10023, inplace=True)
trees_raw_df['zipcode'].replace(10111,10001, inplace=True)
trees_raw_df['zipcode'].replace(10115,10027, inplace=True)
trees_raw_df['zipcode'].replace(10281,10282, inplace=True)
trees_raw_df['zipcode'].replace(11096,11691, inplace=True)
trees_raw_df['zipcode'].replace(11249,11222, inplace=True)
trees_raw_df['zipcode'].replace(11251,11205, inplace=True)
trees_raw_df['zipcode'].replace(11360,11357, inplace=True)
trees_raw_df['zipcode'].replace(11371,11369, inplace=True)
trees_raw_df['zipcode'].replace(11451,11433, inplace=True)
trees_df = pd.merge(trees_raw_df, converter_df, left_on="zipcode", right_on="Zip Code",how="left", sort=False)
trees_df = trees_df.drop(columns=['created_at','tree_id','block_id','the_geom','boroname','stump_diam','curb_loc','spc_latin','spc_common','steward','guards','sidewalk','user_type','problems','root_stone','root_grate','root_other','trnk_wire','trnk_light','trnk_other','brnch_ligh','brnch_shoe','brnch_othe','address','zip_city','cb_num','cncldist','st_assem','st_senate','nta','nta_name','boro_ct','state','x_sp','y_sp'])
trees_df.fillna("Dead",inplace=True)
trees_df['health'].replace("Good", 3, inplace=True)
trees_df['health'].replace("Fair", 2, inplace=True)
trees_df['health'].replace("Poor", 1, inplace=True)
trees_df['health'].replace("Dead", 0, inplace=True)
trees_df.drop(columns=['zipcode','borocode', 'status'],inplace=True)
```

```python
trees_borough_df = trees_df.drop(columns=['Latitude','longitude', 'UHF34 Code','UHF34 Neighborhood Name'])
trees_grouped_borough = trees_borough_df.groupby('Borough')
trees_borough_df = trees_grouped_borough.mean()
trees_borough_df['Number of Trees'] = trees_grouped_borough['health'].count()
trees_UHF_df = trees_df.drop(columns=['Latitude','longitude'])
trees_grouped_UHF = trees_UHF_df.groupby('UHF34 Neighborhood Name')
trees_UHF_df = trees_grouped_UHF[['health','tree_dbh']].mean()
trees_UHF_df['Borough'] = trees_grouped_UHF[['Borough']].first()
trees_UHF_df['UHF34 Code'] = trees_grouped_UHF[['UHF34 Code']].first()
trees_UHF_df['Number of Trees'] = trees_grouped_UHF[['health']].count()
```

```python
combined_UHF_df = pd.merge(trees_UHF_df, combined_env_UHF_df, left_on="UHF34 Code", right_on="GeoID",how="inner", sort=False)
combined_borough_df = pd.merge(trees_borough_df, combined_env_borough_df, left_on="Borough", right_on="Borough",how="inner", sort=False)
combined_UHF_df.rename(columns={"health": "Average Status","tree_dbh": "Average Diameter",  "Degrees Fahrenheit": "Average Summer Surface Temperature", "Mean mcg/m3": "Average Fine Particle Pollution", "Mean ppb": "Average Ozone Pollution"}, inplace=True)
combined_borough_df.rename(columns={"tree_dbh": "Average Diameter", "health": "Average Status", "Degrees Fahrenheit": "Average Summer Surface Temperature", "Mean mcg/m3": "Average Fine Particle Pollution", "Mean ppb": "Average Ozone Pollution"}, inplace=True)
trees_df.rename(columns={"tree_dbh": "Diameter", "health": "Status", "Latitude":"Lat", "longitude": "Long"}, inplace=True)
```
